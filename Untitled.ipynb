```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, KFold
```

```python
from sklearn.linear_model import LinearRegression

```

```python
df=pd.read_csv('demopc.csv')
df
```

```python
encoding = {'No': 0, 'Yes': 1}
df['Public_Transport_Access'] = df['Public_Transport_Access'].map(encoding)
df
```

```python

df['Recent_Renovation'] = df['Recent_Renovation'].map(encoding)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df=scaler.fit_transform(df[['Area (sq ft)','Bedrooms'	,'Bathrooms',	'Age_of_Property (Years)',	'Nearby_Schools'	,'Nearby_Hospitals',	'Public_Transport_Access',	'Recent_Renovation','Selling_Price (USD)']])
df
```

```python
df2=pd.DataFrame(df)
df2
```

```python
#Dividing the dataset into X and Y
X=df[[0,1,2,3,4,5,6,7]]
Y=df[[8]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
LR=LinearRegression()
LR=LR.fit(X_train,y_train)
print(LR.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
y_pred = LR.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae= mean_absolute_error(y_test, y_pred)
print(f'The Mean Squared Error is: {mse}\nAnd the Mean Absolute Error is: {mae}')
```
